```python
import pandas as pd
import re
import json
```

```python
with open('data/dataYiddish/yiddishland/yiddishland') as f:
    lines = f.readlines()[3:]
```

```python
triplets = [line.split() for line in lines if len(line.split()) == 3]
```

```python
string = triplets[9][0]
```

```python
string
```

```python
pairs = []
```

```python
for line in triplets:
    title = re.sub('[^\u0590-\u05fe \-]+', '', line[2])
    trans = re.sub('[^\w -]', '', line[0]).lower()
    parallels = list(zip(title.split(), trans.split()))
    for parallel in parallels:
        print(parallel)
        pairs.append(parallel)
```

```python
len(pairs)
```

```python
pair_df = pd.DataFrame({'yiddish': [tup[0] for tup in pairs], 'translit': [tup[1] for tup in pairs]})
```

```python
pair_df['source'] = 'yivo'
```

```python
len(pair_df), len(pair_df.drop_duplicates())
```

```python
pair_df.drop_duplicates().to_csv('data/processed-parallel/yiddishland.csv', index=False)
```
