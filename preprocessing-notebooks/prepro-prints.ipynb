```python
hebstr = 'אבגדהוזטימנלספערתםןף'
latstr = 'fnmsrepslnmitzvhdgba'

letterdict = defaultdict(str)

for k, v in zip(hebstr, latstr[::-1]):
    letterdict[k] = v

letterdict['ח'] = 'kh' 
letterdict['כ'] = 'kh'
letterdict['ך'] = 'kh'
letterdict['צ'] = 'ts'
letterdict['ץ'] = 'ts'
letterdict['ש'] = 'sh'
letterdict['-'] = '-'
letterdict[' '] = ' '
letterdict['אָ'] = 'o'
letterdict['אַ'] = 'a'
letterdict['פֿ'] = 'f'
letterdict['ײַ'] = 'ay'
letterdict['בֿ'] = 'v'
letterdict["'"] = "'"

for num in range(10):
    letterdict[str(num)] = str(num)
```

```python
matches = defaultdict(lambda: ('', 0),)

hebs = [word[0] for word in hebrew_types]
lats = [word[0] for word in latin_types]

for i, heb in enumerate(hebs):
    if 'א' in heb:
        boundary = 65
    else:
        boundary = 75
    pseudolat = ''.join([letterdict[str(letter)] for letter in re.sub('\W', '', heb)]) 
    ratios = []
    indices = [i] + [i+i_ for i_ in range(1,151)] + [i-i_ for i_ in range(1,151)]
    choices = [lats[i_].lower() for i_ in indices if i_ >= 0 and i_ < len(lats)]
    choices_sans = [re.sub('\W', '', choice) for choice in choices]
    choice_dict = dict(zip(choices_sans, choices))
    choices = [choice for choice in choices_sans if not len(choice)*1.5 < len(pseudolat) and not len(pseudolat)*1.5 < len(choice)]
    best = process.extractOne(pseudolat, choices)
#    print(best)
    if best != None:
        if matches[choice_dict[best[0]]][1] >= best[1]:
            pass
        elif best[1] > boundary:
            matches[choice_dict[best[0]]] = (heb, best[1])
```

```python
with open('prints-latwords.txt') as f:
    latwords = f.readlines()
```

```python
latwords[0][:-1]
```

```python
labels = set([word[:-1] for word in latwords]) - set(labeled.translit.values)
```

```python
len(labels)
```

```python
matches = defaultdict(lambda: ('', 0),)

def best_match(heb):
    if 'א' in heb:
        boundary = 70
    else:
        boundary = 80
    pseudolat = ''.join([letterdict[str(letter)] for letter in re.sub('\W', '', heb)]) 
    ratios = []
    choices = [label.lower() for label in labels]
    choices_sans = [re.sub('\W', '', choice) for choice in choices]
    choice_dict = dict(zip(choices_sans, choices))
    choices = [choice for choice in choices_sans if not len(choice)*1.5 < len(pseudolat) and not len(pseudolat)*1.5 < len(choice)]
    best = process.extractOne(pseudolat, choices)
#    print(best)
    if best != None:
        if matches[choice_dict[best[0]]][1] >= best[1]:
            pass
        elif best[1] > boundary:
            matches[choice_dict[best[0]]] = (heb, best[1])
```

```python
unlabeled.yiddish.apply(lambda x: best_match(x))
```

```python
matches
```

```python
wordlist = [(v[0], k) for k,v in matches.items() if v[1] != 0]
wordlist
```
