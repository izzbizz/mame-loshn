```python
import pandas as pd
import re
```

```python
df = pd.read_csv('data/dataYiddish/songtexts/lider-verter.csv')
```

```python
df.sample()
```

```python
len(df)
```

```python
df['title'] = df.yiddish.apply(lambda x: re.sub('[^\u0590-\u05fe \-]+', '', x))
df['trans'] = df.transliteration.apply(lambda x: re.sub('[^\w \-]+', '', x))
```

```python
df.sample()
```

```python
pairs = []
```

```python
def add_to_pairs(row):
    parallels = list(zip(row.title.split(), row.trans.split()))
    for parallel in parallels:
        print(parallel)
        pairs.append(parallel)
```

```python
df.apply(lambda x: add_to_pairs(x), axis=1)
```

```python
pair_df = pd.DataFrame({'yiddish': [tup[0] for tup in pairs], 'translit': [tup[1] for tup in pairs]})
```

```python
pair_df['source'] = 'yidlid'
```

```python
len(pair_df), len(pair_df.drop_duplicates())
```

```python
pair_df.drop_duplicates().to_csv('data/processed-parallel/songs.csv', index=False)
```
