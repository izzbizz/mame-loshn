```python
import pandas as pd
import re
```

```python
with open('yiddishdict') as f:
    d = f.read()
```

```python
start = d[6000:8000]
```

```python
chapters = d.split('</font></a></b></td></tr><tr><td')
```

```python
split_chapts = [re.split('    +', chapt) for chapt in chapters]
```

```python
[chapt for chapt in split_chapts if len(chapt) != 5]
```

```python
l = []
y = []

for chapt in split_chapts[1:]:
    if len(chapt[0]) > 200:
        l.append(chapt[0])
        y.append(chapt[3])
    else:
        l.append(chapt[1])
        y.append(chapt[4])
```

```python
latin = [re.findall('\>([^\>]+)$', word) for word in l]
yiddish = [re.findall('\>([^\>]+)$', word) for word in y]
```

```python
yiddish = [word for liste in yiddish for word in liste]
translit = [word for liste in latin for word in liste]

yiddish += ['כלאָפּ']
translit += ['khlop']

translit.remove('shlenglen zikh')
translit.remove('shlendern zikh')
```
